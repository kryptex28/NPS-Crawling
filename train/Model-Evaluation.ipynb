```python
from sklearn.model_selection import train_test_split
import pandas as pd
df = pd.read_csv(r"./Mapping_Manuell_V2_Ellena.csv")
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

```

```python
from nps_crawling.classification.model import ClassificationClass, get_model_class
from nps_crawling.classification.options import (
    ClassificationOptionName,
    get_classification_option
)
option = get_classification_option(ClassificationOptionName.NPS_CATEGORY)
model_class = get_model_class(ClassificationClass.SVM, {"model":"BAAI/bge-m3"})
```

```python
for property in option.get_classification_properties():
    model_class.train(classification_property=property, texts=train_df["snippet_text_short"].to_list(), labels=train_df[property.name].to_list())
```

```python
for property in option.get_classification_properties():
    model_class.evaluate(option, property, texts=test_df["snippet_text_short"].to_list(), labels=test_df[property.name].to_list())
```

```python
from nps_crawling.classification.model import ClassificationClass, get_model_class
from nps_crawling.classification.options import (
    ClassificationOptionName,
    get_classification_option
)
option = get_classification_option(ClassificationOptionName.NPS_VALUE)
model_class = get_model_class(ClassificationClass.LLMHUGGINGFACE, {"model":"mistralai/Mistral-7B-Instruct-v0.3"})
```

```python
for property in option.get_classification_properties():
    float_labels = test_df[property.name].apply(lambda x: float(x.replace(',', '.')) if isinstance(x, str) else x).to_list()
    model_class.evaluate(option, property, texts=test_df["snippet_text_short"].to_list(), labels=float_labels)
```
